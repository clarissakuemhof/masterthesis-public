```python
# Load the CSV files (replace with your file paths)
df_eum = pd.read_csv('testing/CAPQL-PGMORL-EUM-hopper.csv')
df_hypervolume = pd.read_csv('testing/CAPQL-PGMORL-HV-hopper.csv')
df_sparsity = pd.read_csv('testing/CAPQL-PGMORL-sparsity-hopper.csv')

df_eum["Step"] = df_eum["Step"] / 1_000_000
df_hypervolume["Step"] = df_hypervolume["Step"] / 1_000_000
df_sparsity["Step"] = df_sparsity["Step"] / 1_000_000

# Create a 1x3 subplot layout (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # Adjusted size for horizontal layout

# EUM Plot (Left) - Only PGMORL
axes[0].plot(df_eum["Step"], df_eum["algo: PGMORL - eval/eum"], label='PGMORL', color='green')
axes[0].fill_between(df_eum["Step"], 
                     df_eum["algo: PGMORL - eval/eum__MIN"], 
                     df_eum["algo: PGMORL - eval/eum__MAX"], 
                     color='green', alpha=0.1)

axes[0].set_title('Expected Utility (↑)')
axes[0].set_xlabel('Training Steps (M)')
axes[0].set_ylabel('EUM')
axes[0].legend()
axes[0].grid(True)

avg_hypervolume = df_hypervolume["algo: PGMORL - eval/hypervolume"].mean()

# Hypervolume Plot (Middle) - Only PGMORL
axes[1].plot(df_hypervolume["Step"], df_hypervolume["algo: PGMORL - eval/hypervolume"], label='PGMORL', color='green')
axes[1].fill_between(df_hypervolume["Step"], 
                     df_hypervolume["algo: PGMORL - eval/hypervolume__MIN"], 
                     df_hypervolume["algo: PGMORL - eval/hypervolume__MAX"], 
                     color='green', alpha=0.1)

axes[1].axhline(y=avg_hypervolume, color='red', linestyle='--', label=f'Average Hypervolume ({avg_hypervolume:.2f})')

axes[1].set_title('Hypervolume (↑)')
axes[1].set_xlabel('Training Steps (M)')
axes[1].set_ylabel('Hypervolume')
axes[1].legend()
axes[1].grid(True)

# Sparsity Plot (Right) - Only PGMORL
axes[2].plot(df_sparsity["Step"], df_sparsity["algo: PGMORL - eval/sparsity"], label='PGMORL', color='green')
axes[2].fill_between(df_sparsity["Step"], 
                     df_sparsity["algo: PGMORL - eval/sparsity__MIN"], 
                     df_sparsity["algo: PGMORL - eval/sparsity__MAX"], 
                     color='green', alpha=0.1)

axes[2].set_title('Sparsity (↓)')
axes[2].set_xlabel('Training Steps (M)')
axes[2].set_ylabel('Sparsity')
axes[2].legend()
axes[2].grid(True)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```
